PCA vs  T-SNE


1)pca is used for the purpose of dimensionality reduction to reduce computational complexity and also to release some zero weighted features while t-sne is only used for visualisation


2)principal components in the pca are the combination of old features with their weightage but there is meaning to the generated components in the t-sne so,pca can handle new points while t-sne cannot


a)pca is linear and t-sne is non linear


b)in the case of non-linearity we will try capture only neighbours so we will get local structure undistorted


c)no t-sne only maintains local structure it doesnot care about global structure




d)t-sne is more computationaly expensive because of perplexity and max_iterations

2)dimensions doesn't really bother but increasing data size bothers as it takes O(n^2) because it need to calculate pair wise probabilities


3)
i)the main thing is tuning hyper parameters we need to check for different values of perplexity and iterations

ii)we loose global structure

iii)computational complexity


```python
import matplotlib.pyplot as plt
import numpy as np;
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
X = digits.data
y = digits.target
X.shape
```

```python
tsne = TSNE(perplexity=40,n_components=2,max_iter=1000)
```

```python
x_embedded = tsne.fit_transform(X)
print("X_embedded shape:", x_embedded.shape)  
print("y shape:", y.shape)                    
```

```python
for label in np.unique(y):
    plt.scatter(
        x_embedded[y == label, 0],
        x_embedded[y == label, 1],
        label=label,
        s=10 
    )

plt.legend(title="Digit")
plt.title('t-SNE on Digits Dataset (2D)')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.grid(True)
plt.show()
```
